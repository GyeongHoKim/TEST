```python
import pandas as pd
from sklearn.preprocessing import scale, robust_scale, minmax_scale, maxabs_scale
from matplotlib import pyplot as plt
import statistics
import numpy as np

df = pd.read_csv('test.csv')

#필요한 값 불러오기
df2 = df[['Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']]

#Sex값 int로 바꾸기
df2_Sex = df2['Sex']
Sex_encoded, Sex_categories = df2_Sex.factorize()
df2['Sex']=Sex_encoded

#Embarked값 int로 바꾸기
df2_Embarked = df2['Embarked']
Embarked_encoded, Embarked_categories = df2_Embarked.factorize()
df2['Embarked']=Embarked_encoded

#Cabin에서 NaN값 중간값으로 대체(4등분해서 계층셈플링)
df2_cabin=df2["Cabin"].tolist()
c=0
for i in df2_cabin:
    i = str(i)
    if i[0]=="A":
        df2_cabin[c]=6.
    elif i[0]=="B":
        df2_cabin[c]=5.
    elif i[0]=="C":
        df2_cabin[c]=4.
    elif i[0]=="D":
        df2_cabin[c]=3.
    elif i[0]=="E":
        df2_cabin[c]=2.
    elif i[0]=="F":
        df2_cabin[c]=1.
    elif i[0]=="G":
        df2_cabin[c]=0.
    else:
        df2_cabin[c]=None
    c=c+1
df2["Cabin"] = df2_cabin
df3 = df2.sort_values(by=['Fare'],axis=0)
Q1 = df3["Cabin"].quantile(0.25)
Q2 = df3["Cabin"].quantile(0.5)
Q3 = df3["Cabin"].quantile(0.75)
Q4 = df3["Cabin"].quantile(1)
Q1_list=[]
Q2_list=[]
Q3_list=[]
Q4_list=[]
for index, row in df2.iterrows():
    if (row["Cabin"] <= Q1):
        Q1_list.append(row["Cabin"])
    elif (row["Cabin"] <= Q2):
        Q2_list.append(row["Cabin"])
    elif (row["Cabin"] <= Q3):
        Q3_list.append(row["Cabin"])
    elif (row["Cabin"] <= Q4):
        Q4_list.append(row["Cabin"])
Q1_median = statistics.median(Q1_list)
Q2_median = statistics.median(Q2_list)
Q3_median = statistics.median(Q3_list)
Q4_median = statistics.median(Q4_list)
df2["Cabin"].fillna(7., inplace=True)
a = len(df2["Cabin"])/4
for index, row in df2.iterrows():
    if (row["Cabin"] == 7.) & (index <= a):
        df2.loc[index, 'Cabin'] = Q1_median
    elif (row["Cabin"] == 7.) & (index <= a*2):
        df2.loc[index, 'Cabin'] = Q2_median
    elif (row["Cabin"] == 7.) & (index <= a*3):
        df2.loc[index, 'Cabin'] = Q3_median
    elif (row["Cabin"] == 7.):
        df2.loc[index, 'Cabin'] = Q4_median

#Age에서 NaN값 중간값으로 대체(Name에서 호칭으로 계층샘플링)
df2['Name'] = df2['Name'].map(lambda x: x.split(',')[1].split('.')[0].strip())
titles = df2['Name'].unique()
df2["Age"].fillna(-1, inplace=True)
medians = dict()
for title in titles:
    median = df2.Age[(df2["Age"] != -1) & (df2["Name"] == title)].median()
    medians[title] = median
for index, row in df2.iterrows():
    if row['Age'] == -1:
        df2.loc[index, 'Age'] = medians[row['Name']]
df2["Age"].fillna(df2["Age"].median(), inplace=True)
df2["Fare"].fillna(df2["Fare"].median(), inplace=True)
df2.drop("Name", axis=1, inplace=True)
df2.drop("Embarked", axis=1, inplace=True)

df2["Pclass"]=minmax_scale(df2["Pclass"])
df2["Sex"]=minmax_scale(df2["Sex"])
df2["Age"]=minmax_scale(df2["Age"])
df2["SibSp"]=minmax_scale(df2["SibSp"])
df2["Parch"]=minmax_scale(df2["Parch"])
df2["Fare"]=minmax_scale(df2["Fare"])
df2["Cabin"]=minmax_scale(df2["Cabin"])
df2
```

```python
import tensorflow as tf
import numpy as np
from functools import partial

x_data = np.loadtxt("TEST_data.csv", delimiter=',', dtype=np.float32)
saver = tf.train.Saver()
n_inputs = 7
n_hidden1 = 300
n_hidden2 = 300
n_hidden3 = 300
n_hidden4 = 300
n_hidden5 = 300
n_hidden6 = 300
n_hidden7 = 300
n_hidden8 = 300
n_hidden9 = 300
n_outputs = 1
learning_rate = 0.01
n_epochs = 2501

with tf.name_scope("prepare"):
    X = tf.placeholder(tf.float32, shape=(None, n_inputs), name = "X")
    Y = tf.placeholder(tf.float32, shape=(None, 1), name = "Y")
    my_dense_layer = partial(
        tf.layers.dense, activation=tf.nn.relu,
        kernel_regularizer=tf.contrib.layers.l2_regularizer(0.001))

with tf.name_scope("dnn"):
    hidden1 = my_dense_layer(X, n_hidden1, name="hidden1", activation = tf.nn.relu)
    hidden2 = my_dense_layer(hidden1, n_hidden2, name="hidden2")
    hidden3 = my_dense_layer(hidden2, n_hidden3, name="hidden3")
    hidden4 = my_dense_layer(hidden3, n_hidden4, name="hidden4")
    hidden5 = my_dense_layer(hidden4, n_hidden5, name="hidden5")
    hidden6 = my_dense_layer(hidden5, n_hidden6, name="hidden6")
    hidden7 = my_dense_layer(hidden6, n_hidden7, name="hidden7")
    hidden8 = my_dense_layer(hidden7, n_hidden8, name="hidden8")
    hidden9 = my_dense_layer(hidden8, n_hidden9, name="hidden9")
    logits = my_dense_layer(hidden9, n_outputs, activation = None, name="outputs")
    
with tf.name_scope("loss"):
    xentropy = tf.nn.sigmoid_cross_entropy_with_logits(labels = Y, logits = logits)
    loss = tf.reduce_mean(xentropy, name="loss")

with tf.name_scope("train"):
    optimizer = tf.train.RMSPropOptimizer(learning_rate)
    training_op = optimizer.minimize(loss)
    
with tf.name_scope("eval"):
    predicted = tf.nn.sigmoid(logits)
    correct = tf.equal(tf.round(predicted), Y)
    accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

with tf.Session() as sess:
    saver.restore(sess, "./my_model_final.ckpt")
    
```
